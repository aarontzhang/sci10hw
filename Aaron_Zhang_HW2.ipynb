## *E. coli* Revisited

Recall that *E. coli* is a _species_ of bacteria. Although we often think of species as being the smallest taxonomic unit (above it is genus and above that are family, order, class, phylum, kingdom, and domain), in bacteria there are often _strains_ within a species. Many strains of *E. coli* are beneficial for our digestive systems; these are called **commensal** strains. But some strains are harmful, which we call **pathogenic** strains. (Equivalently, we say that those strains are **pathogens**.) Antibiotics are often used to treat those pathogenic strains, but *E. coli* has evolved resistance to some antibiotics. We'll use Python to explore that in a moment!

### Antibiotic Resistance and GC content

The antibiotic tetracycline was discovered as a natural product in soil samples by Benjamin Duggar in 1948. It was patented in 1953. Within one year, strains of *Shigella dysenteriae* were identified that already were resistant to this drug. Many bacteria, including pathogenic strains of *E. coli*, have since evolved resistance to tetracycline.

One mechanism of tetracycline resistance arises from the tetA gene. This gene encodes a protein that pumps tetracycline out of the bacterium.

The tetA gene is a sequence of about 1200 A, T, C, and G bases.

You might expect that each of A, T, C, and G would occur with frequency 25%, but actually it's often not the case - in some genes the frequency of bases is different. Moreover, it turns out that the genes like tetA have particularly high abundance of G's and C's. _Why is that?!_ We'll explore this more in the near future!

But, next, we'll write some Python code to see this first-hand.

## `for` loops

In Lab 2, we saw a shiny new feature called **for loops**. This is a nice way to tell Python to repeat something. To remind ourselves of how this works, let's first see the "painful" way of printing the letters in the string "nutella". Run this cell to see that it prints each letter, one-by-one.

```python
def printLetters1():
  print("n")
  print("u")
  print("t")
  print("e")
  print("l")
  print("l")
  print("a")

printLetters1()
```

Now, here's the exact same thing, but this time we're putting each letter in a box (which we've named `letter`) and then printing the content of the box. (This seems like it's only making things more complicated, but humor us for a moment!)

```python
def printLetters2():
  letter = "n"
  print(letter)
  letter = "u"
  print(letter)
  letter = "t"
  print(letter)
  letter = "e"
  print(letter)
  letter = "l"
  print(letter)
  letter = "a"
  print(letter)

printLetters2()
```

Finally, here's a function that does the exact same thing. But, it's much shorter! The `for` loop is automatically placing the symbols in the string `"nutella"` into the box that we've called `letter`. It does that one at a time. So, it first puts `"n"` in the box called `letter` just like in line 2 of the `printLetters2` function in the code cell above. Then, it uses `print` to print the contents of that box, just like in line 3 of the `printLetters2` function above.

Next, it puts `"u"` in the `letter` box just like in line 4 of the `printLetters2` function in the code cell above. Then, it uses `print` to print the contents of that box, just like in line 5 of the of the `printLetters2` function above.

It does this until all the way through the last letter, which is `"a"` in this example.

The _takeaway_ is that the `for` loop allows us to automate the long sequence of instructions like in the `printLetters2`. You can read line 2 in the code cell below as...

<font color = "purple">**for**</font> each letter <font color = "blue">**in**</font> the string `"nutella"` do the stuff in the indented lines that follow. In other words, the 'for' loop is iterating over each letter in the string. Each letter of the string, one-by-one, is placed into a box called 'i'.  We could have picked any other name or symbol for this iterator box, but we will stick to `i`.

```python
def printLetters3():
  for i in "nutella":
    print(i)

printLetters3()
```

## Better yet!

Now, imagine that we want to print each each letter in the string, but we want that string to be _anything_, not just `"nutella"`. We can do this by adding an input variable (aka box) that we specify when calling the function.

The `printLetters4` function below takes a box (aka "variable") as input. We've called it `myFave`.

Now, when we run `printLetters4` function on line 5, the string `"nutella"` goes into the box called `myFave`.

Since `myFave` contains `"nutella"`, line 2 in `printLetters4` does the same thing as line 2 in the code cell above for `printLetters3`.

But, we can change this input! When we run the function _again_ on line 6, the string `"One-eyed aliens rock"` goes into the box called `myFave`. So, line 2 below takes the letters in that string (`"One-eyed aliens rock"`), one-by-one, and puts them into the box called `i`. **So cool!**

```python
def printLetters4(myFave):
  for i in myFave:
    print(i)

printLetters4(myFave = "nutella")
printLetters4(myFave = "One-eyed aliens rock!")
```

## Back to G's and C's!

The code below does something even more interesting! It takes in a DNA string of any length we wish. The `for` loop in line 2 then marches through that string one symbol (letter) at a time. It puts the current symbol in a variable (aka a "box") that we've called `i` (it could have been given any other name). Then, it enters the indented block beginning at line 3 and checks if that symbol is a `"G"`, a `"C"`, or something else. Try it out!

```python
def GChappy(DNAstring):
  for i in DNAstring:
    if i == "G":
      print("Found a G")
    elif i == "C":
      print("Found a C")
    else:
      print("!")

GChappy(DNAstring = "GGACT")
```

Recall from lab 2 that the turtle package has functions `turtle.forward`, `turtle.left`, and `turtle.right` to make the turtle go forward some specified amount, turn left some specified number of degrees, and turn right some specified number of degrees. For example, here's a function that draws a square with a given side length. After you run this once, change the `100` to `200` on line 12. That will draw a square that has a side length of 200 rather than 100.

```python
def square(sideLength):
  turtle.forward(sideLength)
  turtle.right(90)
  turtle.forward(sideLength)
  turtle.right(90)
  turtle.forward(sideLength)
  turtle.right(90)
  turtle.forward(sideLength)
  turtle.right(90)

turtle.clearscreen()
square(sideLength = 100)
```

That function was kind of messy. It repeated the same process (forward, right) four times. The code cell below is a nicer way of doing the same thing. This time, we're calling the function twice, first to draw a square with side length 100 and then to draw one with side length 200.

```python
def square1(sideLength):
  for i in [1, 2, 3, 4]:
    turtle.forward(sideLength)
    turtle.right(90)

turtle.clearscreen()
square1(sideLength = 100)
square1(sideLength = 200)
```

In the code cell above, notice that the important thing about the list `[1, 2, 3, 4]` is that it has four things in it. That meant that the loop was repeated four times: The first time, the number 1 was put into the box that we called `i`. The second time, the number 2 was put into the `i` box, etc.   

It wasn't particularly important that those four things were the numbers 1, 2, 3, 4. They could have been 0, 1, 2, 3 or, for that matter, 42, 42, 42, 42. In fact, we could have put any four things in that list as you can see below.


```python
def square1(sideLength):
  for i in [42, 42, 42, 42]:
    turtle.forward(sideLength)
    turtle.right(90)

turtle.clearscreen()
square1(sideLength = 100)
square1(sideLength = 200)
```

## Functions can call other functions for help...

One of the most powerful features of programs is that one function can "call" another function for "help". Here's an example. Read it first. Try to predict what it will do.
```python
def square1(sideLength):
  for i in list(range(4)):
    turtle.forward(sideLength)
    turtle.right(90)

def prettyDrawing():
  for i in [50, 70, 90, 110]:
    square1(sideLength = i)
    turtle.right(90)

turtle.clearscreen()
prettyDrawing()
```

Notice that the `prettyDrawing` function is _calling_ the `square1` function for help to draw a square of the given side length. Since line 8 is inside a for loop, that function is being called four times!
