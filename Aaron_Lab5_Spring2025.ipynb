## Visualizing the means

In this section, we will use a plot type that we introduced in the previous two labs: the **point plot with error bars**. This type of plot is useful for comparing the means of multiple groups of numerical data.

In a point plot with error bars, the groups that you want to compare are generally on the horizontal axis (as the **independent variable**).  The means of the numerical data are on the vertical axis (as the **dependent variable**).

A dot or other type of "marker" is used to indicate each mean. Error bars are then drawn vertically away from that dot, with the length of the bars corresponding to the standard error for that mean.

In the code cell below, we have a new function called `pointPlotErrorBars`. This function includes three **arguments** to maximize its flexibility:

1. `groupList` -- a list containing the names of the groups
2. `dataList` -- a list containing the numerical data from all of the groups
3. `colorList` -- a list of colors for the dots and error bars (one color per group)

Don't worry too much about the syntax for creating this plot. With just a few minutes on Google, you could find the code to make just about any type of plot that you might want. (Prof. N used this exact strategy when writing this function!)

Run the code cell below to see the resulting point plot with error bars.

```python
import matplotlib.pyplot as plt
import statistics
import math
%config InlineBackend.figure_format='retina'

def SEM(myData):
  sampleMean = statistics.mean(myData)  # calculate mean (and put it in a box)
  stDev = statistics.stdev(myData)  # calculate std. dev. (and put it in a box)
  numObs = len(myData)  # determine n (and put it in a box)
  stdErr = stDev/math.sqrt(numObs)  # calculate the SEM (and put in a box)
  return stdErr

def pointPlotErrorBars(groupList, dataList, colorList):
  numGroups = len(groupList)  # calculate how many groups
  for i in range(numGroups):
    # draw dots at the means
    # and draw error bars centered on the dots
    plt.errorbar(groupList[i], statistics.mean(dataList[i]), yerr = SEM(dataList[i]), fmt = 'o', color=colorList[i])
  plt.xlim(-0.5,1.5)
  # create axes labels and title

  plt.xlabel("Treatment group")
  plt.ylabel("Hair growth (mm)")
  plt.title("Procyanidin B2 Clinical Trial Results")

# lists containing group names, numerical data, and dot/bar colors;
# note that the numerical data is input as a list of lists (one list per group!)
# the order of the lists matches the order of the group names (the placebo list is first in this example)
groups = ['placebo', 'procyanidin']
hair_growth = [[0.3, 1.4, 3, 3.7, -1.5, -2, 0, 4.8, 2.4, -11.3],
              [3.5, 5, 7.3, 18.3, 14.5, 6.7, 9, -0.7, 7.8, -4, 6, 4.5, 8, 11.4, 1, 7.3, 8.5, -0.7, 13.5]]
colors = ['Plum', 'SpringGreen']

pointPlotErrorBars(groups, hair_growth, colors)
```

## Visualizing the distributions of the data

In this section, we will use a plot type that we introduced in the previous lab: the **box plot**. This type of plot is useful for comparing the distributions of multiple groups of numerical data.

The basic layout of the box plot is the same as the point plot with error bars. The groups that you want to compare are generally on the horizontal axis (as the **independent variable**).  The measured values are on the vertical axis (as the **dependent variable**).   

Every box plot captures five essential features of a dataset:

* The minimum value, excluding any _outliers_
* The first quartile (a.k.a. $Q_1$ or 25th percentile)
* The median (a.k.a. $Q_2$ or 50th percentile)
* The third quartile (a.k.a. $Q_3$ or 75th percentile)
* The maximum value, excluding any _outliers_

The "box" for each dataset is drawn from the first quartile to the third quartile. The median is indicated by a line through the box. "Whiskers" are used to indicate the minimum and maximum values of the dataset.

In a box plot, outliers are defined as observations that are more than $1.5 \times (Q_3 - Q_1)$ below the first quartile or $1.5 \times (Q_3 - Q_1)$ above the third quartile. (This is an entirely arbitrary choice made by humans!) They are indicated with dots or some other symbol.

In the code cell below, we have a new function called `boxPlot`. As with the `pointPlotErrorBars` function, this function includes three **arguments** to maximize its flexibility:

1. `groupList` -- a list containing the names of the groups
2. `dataList` -- a list containing the numerical data from all of the groups
3. `colorList` -- a list of colors for filling in the boxes (one color per group)

Run the code cell below to see the resulting box plot.

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'

def boxPlot(groupList, dataList, colorList):
  # create the box plot object
  myPlot = plt.boxplot(dataList, patch_artist=True, tick_labels=groupList)
  # fill boxes with colors (don't worry about this syntax!)
  for patch, color in zip(myPlot['boxes'], colorList):
    patch.set_facecolor(color)
  # draw the medians as black lines (don't worry about this syntax!)
  plt.setp(myPlot['medians'], color='Black')
  # create axes labels and title
  plt.xlabel('Treatment group')
  plt.ylabel('Hair growth (mm)')
  plt.title('Procyanidin B2 Clinical Trial Results')

# lists containing group names, numerical data, and box colors;
# note that the numerical data is input as a list of lists (one list per group!)
# the order of the lists matches the order of the group names (the placebo list is first in this example)
groups = ['placebo', 'procyanidin']
hair_growth = [[0.3, 1.4, 3, 3.7, -1.5, -2, 0, 4.8, 2.4, -11.3],
              [3.5, 5, 7.3, 18.3, 14.5, 6.7, 9, -0.7, 7.8, -4, 6, 4.5, 8, 11.4, 1, 7.3, 8.5, -0.7, 13.5]]
colors = ['Thistle', 'PaleGreen']

# call the boxPlot function!
boxPlot(groups, hair_growth, colors)
```
