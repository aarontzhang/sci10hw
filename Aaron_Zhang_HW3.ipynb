## Population Models and Competition ü§º‚Äç‚ôÄÔ∏è

In class, we explored two common models for population growth: the exponential model and the logistic model. Let's briefly review these models.

In the exponential model, there is a single parameter called `rate` which dictates the rate of change in the population from one time step to the next. (The length of time in one time step is implicit; it may be in minutes for bacteria to years for mammals.)

In the exponential model we have:
<pre>
nextPopSize = currentPopSize + rate * currentPopSize
</pre>

In the logistic model, in addition to the `rate` parameter, we have a second parameter called `capacity` which represents the carrying capacity - the maximum population size that can be supported by the environment. In this model we have:
<pre>
nextPopSize = currentPopSize + rate * currentPopSize * (1 - currentPopSize/capacity)
</pre>
Note that if the carrying `capacity` is very large, the term `currentPopSize/capacity` is very small - perhaps quite close to 0. In that case, the logistic model looks nearly identical to the exponential model. But, as the `currentPopSize` gets close to the carrying `capacity`, the term `currentPopSize/capacity` approaches 1 and thus the term `(1- currentPopSize/capacity)` approaches 0. In that case, `nextPopSize` becomes `currentPopSize` - that is, when the `currentPopSize` gets to the carrying `capacity`, the population size no longer changes from one time step to the next.

Take moment to look through the code below and make sure that you understand how it works, since you'll next be writing some code for a related model - the Lotka-Volterra Model!



```python
import matplotlib.pyplot as plt

# Takes as input the initial population size (startSize), the growth rate (rate),
# the carrying capacity (capacity), and the number of generations to simulate
# (numGenerations) and returns a list of the population sizes at each time step.
def logisticModel(startSize, rate, capacity, numGenerations):
  currentPopSize = startSize
  popSizeList = [currentPopSize]    # Start the popSizeList with just the current size
  for generation in list(range(numGenerations)):
    nextPopSize = currentPopSize + rate * currentPopSize * (1 - currentPopSize/capacity)
    popSizeList = popSizeList + [nextPopSize] # Add the next population size to the list
    currentPopSize = nextPopSize # Make the currentPopSize equal to the nextPopSize
  return popSizeList

# plot logistic curve
popSizeList = logisticModel(startSize = 10, rate = 0.2, capacity = 100, numGenerations = 50)

plt.title("Bacterial Growth Data Using Logistic Model")
plt.xlabel("Generation")
plt.ylabel("Population size")
plt.plot(popSizeList, color = "LimeGreen")

plt.legend() # This draws the legend for the figure
plt.show()   # This says, "show me the figure!"
```

In class, we also saw the Lotka-Volterra model for pairs of species, such as two types of bacteria that are competing for resources or for bacteria and bacteriophages, which are in a constant "arms race" with one another.

The Lotka-Volterrra model is very similar to the logistic model, except that we now have two populations and, in addition to the rate and capacity parameters for _each_ population, we also have a term for the "effect" that species 2 has on species 1 and a term for the "effect" that species 1 has on species 2. If these effect terms are 0, then two species have no effect on one another and the population growth for each species is exactly the logistic model. But, if, for example, the effect of species 2 on species 1 (which we'll call `effect2on1`) is greater than 0, that means that species 2 is doing something to suppress the population growth of species 1 (e.g., competing for food with species 1). Similarly, if the effect of species 1 on 2 is greater than 0, then species 1 is inhibiting the population growth of species 2.

So, we'll have two populations, and we'll keep track of their sizes from one time step to the next in lists called `pop1List` and `pop2List`. The relation between the `currentPopSize`'s and the `nextPopSize`'s is now given by:

<pre>
nextPopSize1 = currentPopSize1 + rate1 * currentPopSize1 * (1 - (currentPopSize1 + effect2on1 * currentPopSize2)/capacity1)

nextPopSize2 = currentPopSize2 + rate2 * currentPopSize2 * (1 - (currentPopSize2 + effect1on2 * currentPopSize1)/capacity2)
</pre>

The code cell below implements the Lotka-Volterra model. Note that it returns two things - `pop1List` and `pop2List`. Python is fine returning more than one thing. Your job is to fill in the missing code.

```python
import matplotlib.pyplot as plt

# Implements the Lotka-Volterra Model taking as input, for each of two species,
# the starting size of the population, the rate, the carrying capacity, and the
# effect term on the other species.
# In addition, the function takes as input the number of generations to simulate.
# The function returns two lists representing the populations of the two species.
def lotkaVolterraModel(startSize1, rate1, capacity1, effect1on2,
                       startSize2, rate2, capacity2, effect2on1, numGenerations):
  currentPopSize1 = startSize1
  currentPopSize2 = startSize2
  pop1List = [currentPopSize1]
  pop2List = [currentPopSize2]
  for generation in list(range(numGenerations)):
    # Fill in about 6 lines of code which are to be performed for each
    # iteration of the for loop
    nextPopSize1 = currentPopSize1 + rate1 * currentPopSize1 * (1 - (currentPopSize1 + effect2on1 * currentPopSize2)/capacity1)
    nextPopSize2 = currentPopSize2 + rate2 * currentPopSize2 * (1 - (currentPopSize2 + effect1on2 * currentPopSize1)/capacity2)
    pop1List = pop1List + [nextPopSize1]
    pop2List = pop2List + [nextPopSize2]
    currentPopSize1 = nextPopSize1
    currentPopSize2 = nextPopSize2

  return pop1List, pop2List


# Testing the code here...
pop1, pop2 = lotkaVolterraModel(startSize1 = 10, rate1 = 1.5, capacity1 = 100, effect1on2 = 1.5,
                                startSize2 = 10, rate2 = 1.5, capacity2 = 100, effect2on1 = 0.5, numGenerations = 50)

plt.title("Lotka-Volterra Model")
plt.xlabel("Generation")
plt.ylabel("Population size")
plt.plot(pop1, color = "Green")
plt.plot(pop2, color = "Red")
plt.legend(["Species 1", "Species 2"])
plt.show()
```

Finally, rather than printing the scores for each word, we'd like to construct a list of the scores. That way, we can do interesting things like compute the average word score of the words in a list, etc.

To that end, here's a quick reminder of how such things work. Note the following features of the code below:
   * The function `lengths` is taking an input that we intend to be a list of strings.
   * On line 2, we're creating a variable (a.k.a. "box") called `output` that is an empty list.
   * On line 3, for each word in `wordList` we:
      * find its length and put that length in a "box" called `wordLen` (line 4)
      * append that length (a number) to the end of the `output` list (line 5)
   * Finally, on line 6, we return that list of numbers.

Run the code cell below to see it in action!

```python
def lengths(wordList):
  output = []
  for word in wordList:
    wordLen = len(word)
    output = output + [ wordLen ]
  return output

words = ["hi", "to", "all", "SCI10ers"]
wordLengths = lengths(wordList = words)
print(wordLengths)
```

Now, copy-and-paste your code from Part 13 above into the code cell below. Then, modify your `scoreListOfWords` function so that it **returns** the list of scores for the words. By returning the list of scores, you'll be able to use those scores to compute the average score, standard deviation, and plot the histogram of scores.

Here's a summary of the updates that you'll need to make to your `scoreListOfWords` function.
   * Just as in the `lengths` example above, you'll need to initialize a new list that will initially be empty. You might call that list `listOfScores` (for example).
   * Each time `scoreOneWord` scores a word, rather than printing it you will append it to the `listOfScores`.
   * When the loop is done, you'll return `listOfScores`.

```python
import statistics
import matplotlib.pyplot as plt

# copy-and-paste your oneLetterScore, scoreOneWord, and
# scoreListOfWords functions directly below!

def oneLetterScore(letter):
  if letter in 'aeilnorstu':
    return 1
  elif letter in 'dg':
    return 2
  elif letter in 'bcmp':
    return 3
  elif letter in 'fhvwy':
    return 4
  elif letter in 'k':
    return 5
  elif letter in 'jx':
    return 8
  elif letter in 'qz':
    return 10
  else:
    return 0

def scoreOneWord(myWord):
  score = 0
  # add your code directly below! (about three lines)
  for letter in myWord:
    score += oneLetterScore(letter)

  return score  # return the score

def scoreListOfWords(listOfWords):
  # add your code directly below! (about three lines)
  scores = []
  for word in listOfWords:
      score = scoreOneWord(word)
      print("Word", word, "has score", score)
      scores.append(score)
  return scores

def wordStatsAndHist():
  myWords = ["alien", "amigo", "butterfly", "boba", "coffee", "donuts", "nutella"]
  listOfScores = scoreListOfWords(listOfWords = myWords)
  print("The average word score is:", statistics.mean(listOfScores))
  print("The std. dev. of scores is:", statistics.stdev(listOfScores))
  plt.hist(listOfScores, color="Chartreuse")  # a good retina-searing color!
  plt.xlabel('Word score')
  plt.ylabel('Count')
  plt.show()

wordStatsAndHist()
```
